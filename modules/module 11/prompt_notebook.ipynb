```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 72
from IPython.display import Markdown
import numpy as np
import os

import openai
import IPython
from langchain_community.llms import OpenAI
from dotenv import load_dotenv
```

```python
# load_dotenv('../teachingk ey.env')

# API configuration
Client = openai.OpenAI(api_key=os.environ.get('OPENAI_API_KEY', '<your OpenAI API key if not set as env var>'))
```

```python
def set_openai_params(model='gpt-3.5-turbo', temp=0.7, max_tokens=256, top_p=1, frequency_penalty=0, presence_penalty=0):
    openai_params = {
        'model': model,
        'temperature': temp,
        'max_tokens': max_tokens,
        'top_p': top_p,
        'frequency_penalty': frequency_penalty,
        'presence_penalty': presence_penalty
    }
    return openai_params

def get_completion(params, messages):
    # GET completion from OpenAI API
    response = Client.chat.completions.create(
        model = params['model'],
        messages = messages,
        temperature = params['temperature'],
        max_tokens = params['max_tokens'],
        top_p = params['top_p'],
        frequency_penalty = params['frequency_penalty'],
        presence_penalty = params['presence_penalty'],
    )
    return response
```

```python
# a basic example
params = set_openai_params()
prompt = 'The sky is'

messages = [{"role":"user", "content":prompt}]
response = get_completion(params, messages)
Markdown(response.choices[0].message.content)
```

```python
# self-consistency
question = "What are some good activities to relax after a long day?"

# generate multiple responses for self-consistency
responses = []
for _ in range(5):  # 5 responses for consistency
    messages = [{"role": "user", "content": question}]
    response = get_completion(params, messages)
    responses.append(response.choices[0].message.content.strip())

for idx, resp in enumerate(responses, start=1):
    print(f"Response {idx}:\n {resp}")

# select the most frequent answer
from collections import Counter
most_common_answer = Counter(responses).most_common(1)[0][0]

print("\n  ** The most consistent answer:\n", most_common_answer)
```

```python
# generated knowledge prompting
question = "Why do planets orbit the sun?"

# generated knowledge steps
knowledge_prompts = [
    "Explain the concept of gravitational force.",
    "Describe how gravity affects objects in space.",
    "Explain how an object's mass influences the gravitational force it exerts.",
    "Describe how gravitational force affects planetary motion around the sun."
]

# collect knowledge statements
knowledge_statements = []
for k_prompt in knowledge_prompts:
    messages = [{"role": "user", "content": k_prompt}]
    response = get_completion(params, messages)
    knowledge_statements.append(response.choices[0].message.content.strip())

# combine the knowledge statements to answer the main question
combined_knowledge = " ".join(knowledge_statements)
final_prompt = f"{combined_knowledge} Now, based on this information, why do planets orbit the sun?"

# generate the final response using the accumulated knowledge
messages = [{"role": "user", "content": final_prompt}]
final_response = get_completion(params, messages)

print("\n  ** Accumulated Knowledge Statements:")
for idx, statement in enumerate(knowledge_statements, start=1):
    print(f"Statement {idx}: {statement}")

print("\n  ** Final Answer:", final_response.choices[0].message.content.strip())
```
