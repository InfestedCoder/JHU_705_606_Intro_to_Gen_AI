```python
Model = pipeline("text-generation", model="meta-llama/Llama-3.2-1B-Instruct", max_length=4096, device=Device)
```

```python
def query(_model, _prompt):
    messages = [{'role':'user', 'content':_prompt}]
    response = _model(messages)
    return response[0]['generated_text'][1]['content']
```

```python
# self-consistency
question = "What are some good activities to relax after a long day?"

# generate multiple responses for self-consistency
responses = []
for _ in range(5):  # 5 responses for consistency
    messages = [{"role": "user", "content": question}]
    response = Model(messages)[0]['generated_text'][1]['content']
    responses.append(response)

for idx, resp in enumerate(responses, start=1):
    print(f"Response {idx}: {resp}")

# select the most frequent answer
from collections import Counter
most_common_answer = Counter(responses).most_common(1)[0][0]

print("\n  ** The most consistent answer: ", most_common_answer)
```

```python
# generated knowledge prompting
question = "Why do planets orbit the sun?"

# generated knowledge steps
knowledge_prompts = [
    "Explain the concept of gravitational force.",
    "Describe how gravity affects objects in space.",
    "Explain how an object's mass influences the gravitational force it exerts.",
    "Describe how gravitational force affects planetary motion around the sun."
]

# collect knowledge statements
knowledge_statements = []
for k_prompt in knowledge_prompts:
    messages = [{"role": "user", "content": k_prompt}]
    response = Model(messages)[0]['generated_text'][1]['content']
    knowledge_statements.append(response)

# combine the knowledge statements to answer the main question
combined_knowledge = " ".join(knowledge_statements)
final_prompt = f"{combined_knowledge} Now, based on this information, why do planets orbit the sun?"

# generate the final response using the accumulated knowledge
messages = [{"role": "user", "content": final_prompt}]
final_response = Model(messages)[0]['generated_text'][1]['content']

print("\n  ** Accumulated Knowledge Statements:")
for idx, statement in enumerate(knowledge_statements, start=1):
    print(f"Statement {idx}: {statement}")

print("\n  ** Final Answer:", final_response)
```
