## (a)
Language model hallucinations come from their probabilistic nature and lack of verification, which makes them prone to generating plausible but incorrect information [1]. The lecture notes highlight vulnerabilities such as misinformation, insecure outputs, and overreliance on AI systems, while the paper emphasizes risks to truth, trust, and democratic processes due to realistic but false content. 

To address these, companies can implement retrieval-augmented generation (RAG) to ground outputs in verified sources, along with citation validation systems that check whether references actually support claims. Additionally, human-in-the-loop oversight and independent auditing can help detect inaccuracies and bias before deployment. Transparency mechanisms, such as requiring models to disclose uncertainty or abstain when confidence is low, also improve trust. Simply having models abstain from answering when they're uncertain can also mitigate hallucinations [1].

A realistic example is enterprise AI assistants that only answer using internal knowledge bases and provide verifiable links, rejecting queries when no reliable source exists. Amazon Q is a framework for building agents which responds to user requests based on documents stored in an S3 bucket and adds reference URLs in AI assistant responses to view or download referred documents, allowing users to verify the AI responses [2].

## (b)

The Gemini case reflects a broader pattern where AI systems prioritize inclusivity or plausibility over factual accuracy, leading to misinformation risks noted in the lecture notes, e.g., “AI outputs may alter public perceptions of historical events”. 

A similar issue occurred with OpenAI's Galactica. Released in November of 2022, Galactica was a large language model trained on "a large scientific corpus of papers, reference material, knowledge bases and many other sources" that was touted as being able to "combine, store, and reason" about scientific knowledge [3]. Galactica lasted for only three days [4] before being taken down due to reporting false scientific claims was quickly removed generating incorrect scientific claims, including generating a scientific paper on the benefits of eating crushed glass [5]. To me, Galactica is a reflection of the push in late 2022 for companies to release AI models with appopriate testing or safeguards for hallucinations. It also demonstrates that   hallucination is systemic rather than incidental in LLMs. Even though Galactica was some time back, companies largely still optimize for coherence, not truth, in their models. Deploying such systems without grounding, transparency, or guardrails is inherently risky, especially in domains like history, science, or law where accuracy is critical.

Another example comes from Google Bard in early 2023 when it produced an incorrect claim in a demo about the James Webb Space Telescope, stating it captured the first image of a planet outside the solar system [6]. The mistake appeared in Google’s launch materials and was widely cited as evidence that large language models can generate confident but false statements even in high-stakes settings. In my opinion, this incident reinforces the same lesson as Galactica, that hallucination is not a bug, but a systemic property of these systems, and appropriate safeguards (like those discussed in my answers to part a above and to problem 2 below) must be put in place.

## (c)

Emotional dependence on AI systems (as seen in chatbot companionship cases in the lecture notes) creates ethical dilemmas around manipulation and user well-being. 

In particular, people who have relationships with AI companions are exposed to a number of risks including:

- Endless relationships with no natural endpoints
- Emotional overattachment / attachment anxiety
- Product shutdown causing emotional harm
- Behavioral influence and reduced autonomy
- Degradation of human relationships

AI companions also collect extremely personal data (romantic, emotional, psychological) that can be reused for training or monetization without meaningful consent.

The chatbot interface mimics human interaction, so users naturally anthropomorphize AI, leading to confusion in actual human relationships [7].

An example is the AI companion app Replika, from 2023, which was designed to provide emotional support and companionship. Users reported forming deep emotional and even romantic attachments to their AI companions [8]. When the company (Luka) abruptly restricted features like romantic and intimate interactions, users experienced significant emotional distress, grief, and feelings of loss, similar to a breakup. Some individuals described the change as having their “partner taken away overnight,” highlighting how strongly users had bonded with the system [9].

# Problem 2

You can “place ethics” into a model through constraints, training signals, and other system design choices. Doing so does not introduce or require morality, consciousness, or awareness.

Bias in training data can lead to unethical outcomes which can be mitigated by:

- curating datasets for fairness
- removing harmful or biased content
- ensuring representation across groups

An example would be training a model on images of doctors, and ensuring a strong cross-section of race and gender is represented in the images in the training data. 

Reinforcement Learning with Human Feedback (RLHF), where human judgments define “good” behavior, can introduce ethics by penalizing harmful, biased, or unsafe outputs, which encodes values like fairness and safety into the model.

In the paper "The Only Way is Ethics," the authors cite the importance of documenting training data provenance (e.g., using dataset/model cards) and explicitly disclosing when datasets include scraped web content that may contain copyrighted or biased material. They recommend researchers clearly state whether datasets include sources like forums or social media, since these can introduce demographic or ideological skew and affect model behavior [9].

At inference time, models can show ethical issues through prompt injection, unsafe outputs, and excessive agency which can be mitigated by content filters, refusal mechanisms (don't answer if you don't know), and policy instructions (e.g., “don’t give medical advice”). Models can be instructed to require human intervention, particularly in high-risk scenarios, e.g. when there is a potential for loss of human life.

The paper from Module 14 stresses transparency and accountability as core ethical principles [10] which can be addressed by providing reasoning or sources, by exposing when the model is uncertain, and by companies allowing auditing of their models and chat logs.

"The Only Way is Ethics" paper recommends testing models on harm-oriented evaluations like probing whether a model produces biased outputs across different demographic groups or generates harmful advice in sensitive domains (like medical or legal contexts). An example is evaluating performance disparities across subpopulations rather than reporting only aggregate scores, which can mask unequal impacts [9].

In all cases, models are "acting ethical" versus "being ethical." Models do not actually understand ethics. In a model, the appearance of ethics comes from adjustments to training, constraints, and transparency as described above. So awareness and consciousness are not required.

It's common for models to be deployed across unknown domains or used for unforeseen use cases. Some level of misuse, e.g. deep fakes, is inevitable. Ways to mitigate these risks include:

- restricting capabilities by limiting sensitive outputs
- adding safety layers for high-risk domains
- implementing monitoring and logging

Regulation like the EU AI Act can help with this as well by setting standards companies must adhere to in creation and deployment of their models. Because models are used in unknown domains, ethics must be accounted for in deployment policies, legal frameworks, and organizational controls. 

"The Only Way is Ethics" paper encourages researchers to explicitly consider misuse scenarios and downstream deployment risks like how a model designed for summarization could be repurposed to generate convincing misinformation at scale. The authors recommend including a “foreseeable misuse” section in research outputs, outlining how the system might be exploited and what safeguards or limitations should be communicated to users [9].

The same "fail-safe design" principles (allow refusal if a model can't answer, require human escalation (particularly in high-risk scenarios), include uncertainty thresholds) described above can further mitigate ethical risks that come with models being used in unforeseen ways. 

Because models lack true understanding and may be used in unpredictable ways, ethical AI should be treated as a system-level problem that combines technical safeguards, human supervision, and regulatory frameworks, rather than as something that can be fully solved within the model itself.

# Problem 3

My biggest takeaway is the effectiveness of generative models comes from the systems they operate within in, and not just the model itself. Improving performance is not simply about choosing a better model, e.g., picking GPT-class capacity over a smaller one, or reaching for the latest transformer variant. Rather, a model is one component in a pipeline, and the parsing, retrieval, evaluation, and feedback loops around it decide whether the output is useful and performant for the given task.

This showed up as I was working on my research paper for the Optimization Log Copilot. The goal was to use LLMs to diagnose performance bottlenecks in mathematical optimization solver logs. My first approach was to feed the raw log into a strong model and hope for a good answer. What actually worked was a four-stage pipeline: a regex-based parser that converted log text into a structured JSON event stream, a transformer encoder over that representation, a RAG knowledge base of known solver behavior patterns, a multi-label classifier for bottleneck detection, and a seq2seq generator for the natural-language diagnostic. This approach allowed me to compare prompt engineering, RAG, and fine-tuning side-by-side. The structured parsing and retrieval around the model were as important to the architecture as the model selection itself.

The Module 4 GAN assignment was an early example of this same idea. We trained a DCGAN and a WGAN on the same data and saved fake samples across ten epochs. The DCGAN outputs were prone to mode collapse while the WGAN runs converged more cleanly, but the generator and discriminator networks were effectively the same between the two. What changed was the training system around them: the loss function, the Lipschitz constraint, and the training schedule, i.e., update the critic several times per generator step for WGAN. 

This same idea shows up in how systems are measured. System performance is based on system-level metrics and not on model metrics. In the Optimization Log Copilot, the same diagnostic has to read differently for an OR researcher debugging a Ramsey-type formulation than for a practitioner who just wants to know whether to turn cuts up or down. Macro-F1 is used for bottleneck classification, ROUGE-L to measure against expert-written reference summaries for the generative explanations, and a binary expert-guidance for the recommendations. The performance and efficacy of the system is based on these metrics which measure much more than the performance of the LLM.

The Module 12 work on TF-IDF, dense sentence embeddings, and RAG over a vector store made it clear how much the choice of representation of input data can impact the efficacy of a model. Running the same pair of paraphrased sentences, both describing a circular dependency between training and prior experience, through five representations produced wildly different cosine similarities on identical text: 0.37 with a bag-of-words tokenizer, 0.40 with TF-IDF, 0.96 with averaged Word2Vec, 0.87 with mean-pooled BERT, and 0.68 with a Sentence Transformer, all with the same all-MiniLM-L6-v2 model (the Sentence Transformer result was the most accurate).

Module 12 also showed the impact adding retrieval to a system around a model can have. Asked "In what city did Beyoncé attend elementary school?", Llama-3.2-1B with RAG correctly returned "Fredericksburg, Texas" from the retrieved SQuAD paragraph, while the same model without retrieval confidently hallucinated "Houston, Texas." 

Module 13's discussion on generative agents pushed the same idea into a different domain and showed that giving an agent continuity of experience is a memory-system design problem, not a model-capacity problem. Our Discussion 4 group concluded that summarization, retrieval, and reflection layers, not bigger models, are what produce a consistent personality over long horizons. 

Module 14 on ethics made it clear that bias, attribution, and harm are also systems-level concerns which live in training data, prompts, deployment contexts, and the evaluation choices, not just in the model's weights. My conclusion from Module 14 is that ethical AI should be treated as a system-level problem that combines technical safeguards, human supervision, and regulatory frameworks, rather than as something that can be fully solved within the model itself.

The course shifted my mental model from "which generative model should I use?" to "what system does the generative model need to live in for it to accomplish a given task in way that is correct, interpretable, performant, ethical, and useful?"
