# Problem 1

## (a)

For the if-side of the expression, we have a conjunction:
$$
\Box B_A \land \Diamond D_A \land \Box I_A
$$

For the then-side, "A tries to fulfill I" can be represented by $I_A$. That happens when B holds, which is $B_A \rightarrow I_A$. And this occurs "in every world," giving:

$$
\Box(B_A \rightarrow I_A).
$$

Putting it all together, we have 

<p style="text-align: center;">if</p>

$$
\Box B_A \land \Diamond D_A \land \Box I_A 
$$

<p style="text-align: center;">then</p>

$$
\Box(B_A \rightarrow I_A).
$$

## (b) 

We want to show 

$$
(\Box B_A \land \Diamond D_A \land \Box I_A)
\rightarrow
\Box(B_A \rightarrow I_A)
$$

**Proof.** Assume:
$$
\Box B_A \land \Diamond D_A \land \Box I_A
$$

By conjunction elimination (https://en.wikipedia.org/wiki/Conjunction_elimination):
$$
\Box B_A,\quad \Diamond D_A,\quad \Box I_A
$$

By the Truth Axiom (T), $\Box \phi \Rightarrow \phi$, we obtain:
$$
B_A \quad \text{from } \Box B_A,\qquad I_A \quad \text{from } \Box I_A
$$

Since $B_A$ and $I_A$ hold, the implication
$$
B_A \rightarrow I_A
$$

also holds by material implication (https://en.wikipedia.org/wiki/Material_conditional). 

$\Box B_A$ and $\Box I_A$ assert that these statements hold in all accessible worlds, so the implication also holds in all such worlds. Therefore:

$$
\Box(B_A \rightarrow I_A)
$$

Thus:
$$
(\Box B_A \land \Diamond D_A \land \Box I_A)
\rightarrow
\Box(B_A \rightarrow I_A)
$$

To address logical omniscience, we do not assume the agent derives all logical consequences of its beliefs. Instead, the implication $B_A \rightarrow I_A$ must be explicitly represented or derivable using a limited set of inference rules. Under this bounded-reasoning assumption, the conclusion holds.

# Problem 2

## (a)

A shared data fusion protocol is a structured way for multiple agents (or models, sensors, simulators) to exchange, combine, and reconcile their information so they agree on a consistent view of the world (https://link.springer.com/chapter/10.1007/978-3-030-61108-8_19). 

Each agent maintains could exist in a BDI framework, maintianing beliefs about local atmospheric conditions, goals for acccurate prediction, and intentions to update and refine predictions.​

Without GPT, we could have the shared data fusion protocol align agents by making each agent publish its local state and uncertainty. Overlapping estimates could be handled through consensus or data assimilation (https://en.wikipedia.org/wiki/Data_assimilation), and the combined state could be fed back into each agent's simulation at every timestep. 

At each timestep, agents would:
- Simulate local dynamics
- Exchange information
- Apply fusion update
- Repeat

We could extend this using ensemble forecasting (https://en.wikipedia.org/wiki/Ensemble_forecasting) which would allow each agent to represent its local state as a distribution rather than a point estimate. We could use the mean and covariance of each ensemble to create the combined state. This would improve robustness, account for uncertainty, and lead to more stable and consistent alignment across shared regions.

For example, assume the system has three agents representing west, central and east regions where each agent predicts temperature and pressure. Each agent would run multiple simulations then share the mean and covrariance estimates at boundaries with the adjoining regions (east shares with central, central shares with east and west, west share with central). In the non-GPT system, all that is shared are the estimates and hard-coded rules figure out how to resolve discrepancies, e.g., consensus or data assimilation.

## (b)

Instead of using hard-coded fusion rules, we could use GPT agents to interpret model outputs, reason about discrepancies, negotiate alignment when there are differences, and summarize regional or other patterns in the system. 

The distributed consensus and data assimilation methods used without GPT could be replaced by agents using shared memory, natural language communication, and dynamic reasoning to align the models and solve inconsistencies. When accessing shared memory, agents could be intelligent about result retrieval taking into account things like recency, importance, and relevance.

At each timestep:
- Agents simulate local dynamics
- Share their results and reasoning
- Discrepancies are identified and addressed by a reasoning agent and agents are adjusted as needed
- Results are summarized and shared out in natural language
- Repeat

In our example with the west, central, and east region agents, each agent could reason about its mean and covariance data and share that reasoning in natural language with the other agents, e.g., the west agent might share, “My temperature spike is due to a simulated pressure drop—possibly overestimated due to low grid resolution." The central agent could take this and compare west's results to both its own results and east's results and apply some reasoning, e.g., "The gradient is unusually steep. West’s model may be overreacting to boundary conditions." 

A reasoning agent could further handle negotiation, alignment, and summarizing the results, e.g, “Reduce West’s influence by 30% due to high variance and suspected bias."

This all comes with trade-offs around system complexity, e.g., we would need to mitigate hallucinations, reduced determinism (the system with GPT is more probabilistic, without GPT it's more deterministic), and likely higher compute costs.



# Problem 3

## (a) 

We could use a Belief–Desire–Intention (BDI) multi-agent framework to model NPCs where each agent maintains beliefs about the world + other agents, desires, and intentions (commited actions). Agents could track relationship scores and a history of interactions, increasing relationship scores when there is collaboration and lowering them when there is betrayal. Hard-coded rules or heuristics would need to be in place around relationship scoring and classification of actions as being collaborative or being a betrayal. Agents would need to maintain persistent goals until they are either achieved or proven impossible to achieve. Reactions to betrayal and collaboration would need to be built into the system. 

By sharing information about their goals, desires, and past interactions with one another, agents could start to form clans or other social groups. Alliance formation could be represented as a graph problem, where agents are nodes and relationship scores are weighted edges. Clans or social groups emerge as clusters of high-trust agents. The limitation of this approach is that it depends heavily on hand-designed rules, scoring functions, and other pre-defined models, e.g., agents could use mathematical optimization-based decision-making using pre-defined mixed-integer programming models to guide their decision-making. 

Agents could also use higher-order reasoning, meaning they could model not only their own beliefs, but also what other agents believe or intend. For example, an NPC might believe that another NPC intends to betray the group, which could affect trust and future cooperation.


## (b) 

With GPTs, agents could reason about their past interactions with other characters, dynamically classify current and past interactions, and generate believable social responses to actions using inference at run-time. Agents would still have beliefs, desires, and goals, but they would be able to reason about those for themselves. Overall, the system with GPTs would be much more dynamic with fewer rules needing to be hard-coded into the system. Social dynamics such as alliances and reactions to betrayal could emerge from a loop of memory, reflection, and communication with other agents.

The system could use a memory stream to store observations, conversations, and past interactions. At each decision point, agents could retrieve relevant memories based on recency, importance, and relevance. Reflecting on these memories could allow agents to summarize their raw memories into higher-level social beliefs such as “I trust this character,” “this character betrayed me,” or “I feel welcomed in this group.”

Agents could have some level of planning to help preserve coherence over time, so agents would not simply react moment-by-moment but pursue longer-term goals and relationships.

The main difference is that the non-GPT approach requires explicit rules for relationship updates, betrayal detection, and alliance formation, while the GPT-based approach can infer these patterns from memory and language context. However, GPT agents may be less predictable and may require safeguards to prevent hallucinated or inconsistent behavior.

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 72
from IPython.display import Markdown
import numpy as np
import os

MODEL_PATH= 'EP_models/'
os.environ['HF_HOME'] = MODEL_PATH  # before import transformers
os.environ['HF_DATASETS_OFFLINE']= '1'
````

````python
import torch
import transformers
from transformers import pipeline

# filter warnings
import warnings
transformers.logging.set_verbosity_error()
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)

Device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

print(f'PyTorch version= {torch.__version__}')
print(f'transformers version= {transformers.__version__}')
print(f'CUDA available= {torch.cuda.is_available()}')
````

````python
# global vars, to clear memory keep them in separate Cells
modelA, modelB = None, None
````

````python
Model_Name_Default='Qwen/Qwen3-4B-Instruct-2507'

# load the tokenizer and model
def get_model(model_name:str):
    return pipeline('text-generation', model=model_name, max_length=4096, device=Device) 

def get_modelA(model_name:str=Model_Name_Default):
    global modelA
    if modelA is not None:  # reclaim GPU memory
        modelA = None
        torch.cuda.empty_cache()
    modelA = get_model(model_name)

def get_modelB(model_name:str=Model_Name_Default):
    global modelB
    if modelB is not None:  # reclaim GPU memory
        modelB = None
        torch.cuda.empty_cache()
    modelB = get_model(model_name)

def prompt(_model, _context, _text, max_length):
    prompt = f"Context: {_context}\n\nQuestion: {_text}\nAnswer:"
    response = _model(prompt, max_length=max_length)
    return response[0]['generated_text']

def promptA(_context, _text, max_length=500):
    global modelA
    return prompt(modelA, _context, _text, max_length)

def promptB(_context, _text, max_length=500):
    global modelB
    return prompt(modelB, _context, _text, max_length)
````

## Qwen/Qwen3-4B-Instruct-2507

Here we try using Qwen/Qwen3-4B-Instruct-2507. The Qwen3 family is particularly good at communication tasks (https://arxiv.org/pdf/2604.10866), even if it doesn't outperform some of the top models like GPT-5.2. 

Using a model with more parameters, 4B over the 1.5B used in the original notebook should result in better reasoning.

````python
get_modelA('Qwen/Qwen3-4B-Instruct-2507')
get_modelB('Qwen/Qwen3-4B-Instruct-2507')
````

````python
prompts = {
    'Alpha': (  # starter text to the tuple
        "You are Agent Alpha. You are talking to Agent Beta. ",
        "Beta, I have mapped out the landing zones. Zone A has a flat terrain suitable for setting up our base, but it lacks water resources. "
        "Zone B has water but is closer to volcanic activity. Which do you think we should prioritize?. Please explain concisely. "
    ),
    'Beta': (  # append more text to the tuple
        "You are Agent Beta. You are talking to Agent Alpha. "
    )}

def gpt_interact():
    global modelA, modelB
    # print model
    print(f'** Using modelA= {modelA.model.name_or_path} modelB= {modelB.model.name_or_path} **\n')
    
    # starter prompt
    responseA = promptA(prompts['Alpha'][0], prompts['Alpha'][1])
    responseA = responseA.split('Answer:')[1].strip()
    display(Markdown('Alpha says: ' + responseA + '\n\n----------------------------------------\n'))
    
    responseB = promptB(prompts['Beta'][0], 'Agent Alpha says: '+responseA)
    responseB = responseB.split('Answer:')[1].strip()
    display(Markdown('Beta says: ' + responseB + '\n\n----------------------------------------\n'))
    
    responseA = promptA(prompts['Alpha'][0], 'Agent Beta says: '+responseB)
    responseA = responseA.split('Answer:')[1].strip()
    display(Markdown('Alpha says: ' + responseA + '\n\n----------------------------------------\n'))
    
    responseB = promptB(prompts['Beta'][0], 'Agent Alpha says: '+responseA)
    responseB = responseB.split('Answer:')[1].strip()
    display(Markdown('Beta says: ' + responseB + '\n\n----------------------------------------\n'))
````

````python
%%time

gpt_interact()
````

## Scheduling with Qwen3-30B-A3B family

Below, I split out the models so we can use Qwen/Qwen2.5-1.5B-Instruct for the agents and Qwen/Qwen3-4B-Thinking-2507 for reasoning. 

````python
from transformers import AutoTokenizer, AutoModelForCausalLM

if modelA is not None:  # reclaim GPU memory
    modelA = None
if modelB is not None:  # reclaim GPU memory
    modelB = None
torch.cuda.empty_cache()

# load model and tokenizer

# --- Agent model ---
agent_model_name = 'Qwen/Qwen2.5-1.5B-Instruct'
agent_tokenizer = AutoTokenizer.from_pretrained(agent_model_name)
agent_model = AutoModelForCausalLM.from_pretrained(agent_model_name, dtype=torch.bfloat16).to(Device)

# --- Reasoning model ---
reasoning_model_name = 'Qwen/Qwen3-4B-Thinking-2507'
reasoning_tokenizer = AutoTokenizer.from_pretrained(reasoning_model_name)
reasoning_model = AutoModelForCausalLM.from_pretrained(reasoning_model_name, dtype=torch.bfloat16).to(Device)
````

````python
%%time

# define prompts for each agent
agent_prompts = {
    "Alpha": (
        "You are Agent Alpha. You prefer mornings (9 AM – 12 PM) and are unavailable on Monday. "
        "Please propose meeting times that fit your schedule."
    ),
    "Beta": (
        "You are Agent Beta. You prefer afternoons (1 PM – 4 PM) and are unavailable on Wednesday. "
        "Please propose meeting times that fit your schedule."
    ),
    "Gamma": (
        "You are Agent Gamma. You prefer late mornings (11 AM – 1 PM) and are unavailable on Tuesday. "
        "Please propose meeting times that fit your schedule."
    )}

# generate responses for each agent
responses = {}
for agent, prompt in agent_prompts.items():
    input_ids = agent_tokenizer.encode(prompt, return_tensors="pt").to(Device)
    output = agent_model.generate(input_ids, max_new_tokens=300)
    responses[agent] = agent_tokenizer.decode(output[0], skip_special_tokens=True)

# combine responses and send them to the reasoning GPT
reasoning_prompt = (
    "You are the Reasoning Agent. The following are responses from three scheduling agents:\n\n"
    f"Alpha: {responses['Alpha']}\n"
    f"Beta: {responses['Beta']}\n"
    f"Gamma: {responses['Gamma']}\n\n"
    "Apply reasoning, including persistent goal logic, to find a common meeting time that satisfies as many preferences as possible. "
    "If no time fully satisfies all preferences, propose the best alternative and explain your reasoning but please be concise. "
)

input_ids_reasoning = reasoning_tokenizer.encode(reasoning_prompt, return_tensors="pt").to(Device)
output_reasoning = reasoning_model.generate(input_ids_reasoning, max_new_tokens=1000)
reasoning_response = reasoning_tokenizer.decode(output_reasoning[0], skip_special_tokens=True)
````

````python
md = "## Individual Agent Responses\n\n"

for agent, response in responses.items():
    md += f"### {agent}\n\n{response}\n\n---\n\n"

display(Markdown(md))
````

````python
print('Reasoning Agent Response:')
Markdown(reasoning_response)
````
