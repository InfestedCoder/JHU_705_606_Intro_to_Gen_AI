## Entropy
Let $\texttt{p}(x)$ be the probability mass function of a random variable $X$, 
over a discrete set of symbols (or __alphabet__) $\mathcal{X}$: $\texttt{p}(x)=P(X{=}x), x \in \mathcal{X}$

Entropy is defined as

$\displaystyle
H(\texttt{p})=H(X)=-\sum\limits_{x\in X}\texttt{p}(x)\log_2\texttt{p}(x)
\qquad \qquad \qquad \text{(1)} $

Entropy measures the amount of information in a random variable. It is usually measured in bits. 
Hence, the log to base 2, but using any other base, yields only a linear scaling of results. 
When entropy is involved, the plain logarithm function `log` should be read as 'log to the base 2'. 
Also, for the sake of this definition, define $0 \log 0 = 0$.

```python
# A log function returning 0 for 0log0 - taking care of an exception
def log2(_x):
    from math import log2
    return 0 if _x==0 else log2(_x)

# x random variable, h entropy
x, h = [], []
for w in np.linspace(0.,1.,20):
    x += [w]
    h += [-w*log2(w)-(1-w)*log2(1-w)]

# plot the relation
_, ax = plt.subplots(1, figsize=(5, 3))
ax.plot(x, h, 'k--', lw=1, label='$H(p)$')
ax.legend()
ax.set(xlabel='$x$', ylabel='Entropy', title='Entropy as a Function of $x$')

plt.show()
```
